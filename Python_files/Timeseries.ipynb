```python
import pandas as pd
import requests
import json
```

```python
timeseries_url = 'https://data.covid19india.org/v4/min/timeseries.min.json'
```

```python
r2 = requests.get(timeseries_url)
timeseries = r2.json()
```

```python
df_timeseries = pd.DataFrame(timeseries)
```

```python
dfAN = pd.DataFrame(df_timeseries['AN']['dates']).transpose().apply(pd.Series)
dfAN.drop(columns = ['delta','delta7'],inplace = True)
AN = dfAN['total'].apply(pd.Series)
AN = AN.fillna(0)
```

```python
AN = AN.drop(columns = 'other')
AN.reset_index(level=0,inplace=True)
AN.rename({'index':'dates'},axis=1,inplace=True)
AN['dates'] = pd.to_datetime(AN['dates'])
```
